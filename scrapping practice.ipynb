```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup ('https://gunacinema.com/','html.parser')
data = requests.get(url)
print(data)
```

```python
source = data.text
lxml_data = BeautifulSoup(source, "lxml")
lxml_data
```

```python
table_code = lxml_data.table
tags = table_code. find_all('tr')

```

```python
data = []
for tag in tags:
    data.append(tag.text.split('\n'))
    
print(data)
```
