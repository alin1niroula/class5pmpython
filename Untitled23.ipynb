```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/gainers', 'html.parser')
data = requests.get(url)
data
```

```python
source = data.text
lxml_data = BeautifulSoup(source, "lxml")
lxml_data
```

```python
table_code = lxml_data.table
tags = table_code.find_all('tr')
```

```python
data = []
for tag in tags:
  data.append(tag.text.split('\n')[1:-1])

data
```

```python
data_stoke = []
for i in data_stoke:
  if i[0] != "":
    data_stoke.append(i)

print(data_stoke)
```

```python
import csv
f= open('data.csv','w')
x= csv.writer(f)
for i in data_stoke:
    x.writerow(i)
```
